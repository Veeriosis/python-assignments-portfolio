```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def gradient_descent(x0, y0, f, grad_f, alpha, num_iterations):
    """
    Performs gradient descent on a function of two variables.
    
    Parameters:
    x0, y0: Initial point for the descent.
    f: Function to minimize.
    grad_f: Function that returns the gradient of f.
    alpha: Learning rate.
    num_iterations: Number of iterations.
    
    Returns:
    (x, y): Coordinates of the final point after gradient descent.
    """
    x, y = x0, y0 
    
    for _ in range(num_iterations):
        grad_x, grad_y = grad_f(x, y)  # Compute gradient
        x -= alpha * grad_x  # Update x in opposite direction of gradient
        y -= alpha * grad_y  # Update y in opposite direction of gradient
    
    return x, y

# Function 1: f1(x, y) = x^2 + y^2
def fun_1(x, y):
    return x**2 + y**2

def grad_f_1(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return grad_x, grad_y

# Test cases for f1(x, y)
x1, y1 = gradient_descent(0.1, 0.1, fun_1, grad_f_1, 0.1, 10)
x2, y2 = gradient_descent(-1, 1, fun_1, grad_f_1, 0.01, 100)
print(f"Final position for (0.1, 0.1): ({x1:.5f}, {y1:.5f})")
print(f"Final position for (-1, 1): ({x2:.5f}, {y2:.5f})")

# Function 2: f2(x, y)
def fun_2(x, y):
    return 1 - np.exp(-x**2 - (y - 2)**2) - 2 * np.exp(-x**2 - (y + 2)**2)

def grad_f_2(x, y):
    grad_x = (2 * x) * (np.exp(-x**2 - (y - 2)**2) + 2 * np.exp(-x**2 - (y + 2)**2))
    grad_y = (2 * (y - 2) * np.exp(-x**2 - (y - 2)**2) + 4 * (y + 2) * np.exp(-x**2 - (y + 2)**2))
    return grad_x, grad_y

# Test cases for f2(x, y)
x3, y3 = gradient_descent(0, 1, fun_2, grad_f_2, 0.01, 10000)
x4, y4 = gradient_descent(0, -1, fun_2, grad_f_2, 0.01, 10000)
print(f"Final position for (0, 1): ({x3:.5f}, {y3:.5f})")
print(f"Final position for (0, -1): ({x4:.5f}, {y4:.5f})")

# f2(x, y)
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(X, Y)
z = fun_2(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("f2(x, y)")
plt.show()
```
